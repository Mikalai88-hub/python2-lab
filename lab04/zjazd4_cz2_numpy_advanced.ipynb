```python
import numpy as np
```

```python
# Ceny 4 produktów w 3 sklepach
ceny = np.array([[29.99, 31.99, 28.99],    # Produkt A
                  [49.99, 52.99, 47.99],    # Produkt B
                  [14.99, 15.99, 13.99],    # Produkt C
                  [99.99, 104.99, 94.99]])  # Produkt D

sklepy = ['Centrum', 'Galeria', 'Online']
produkty = ['A', 'B', 'C', 'D']
print(f"Kształt macierzy: {ceny.shape}")  # (4, 3)
```

Zadanie 1a

```python
ceny_inflacja= ceny* 1.05
print(f'Ceny po inflacji 5% {ceny_inflacja.round(2)}')
```

Zadanie 1b

```python
rabaty = np.array([0.10, 0.05, 0.15])
ceny_promocyjne = ceny * (1 - rabaty)
print(f'Ceny po rabacie {ceny_promocyjne.round(2)}')
```

Zadanie 1c

```python
marza = np.array([[0.30], [0.40], [0.25], [0.50]])
koszt_zakupu = ceny / (1+marza)
print(f'Koszt zakupów bez marzy:{koszt_zakupu.round(2)}')
```

Zadanie 1d

```python
indeksy_najtańszych = ceny.argmin(axis=1)
print('Najtańszy sklep dla każdego produktu:')
for i in range(len(produkty)):
    nazwa_produktu= produkty[i]
    nazwa_sklepu= sklepy[indeksy_najtańszych[i]]
    print(f'Produkt {nazwa_produktu}: {nazwa_sklepu}')
```

Ćwiczenie 3

```python
np.random.seed(2026)

# 8 pracowników, dane roczne
pracownicy = ['Anna', 'Jan', 'Ewa', 'Marek', 'Kasia', 'Piotr', 'Zofia', 'Tomek']
pensje = np.array([5500, 7200, 4800, 9100, 6300, 4200, 8500, 5800])
staz = np.array([3, 8, 2, 15, 5, 1, 12, 4])
oceny = np.array([4.2, 3.8, 4.5, 4.0, 4.7, 3.5, 4.3, 3.9])

# Sprzedaż miesięczna 4 produktów przez 6 miesięcy
produkty_nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Akcesorium']
sprzedaz = np.random.randint(20, 200, size=(4, 6))
ceny_produktow = np.array([3500, 1800, 2500, 150])
```

Zadanie 3a

```python
średnia = pensje.mean()
mediana= np.median(pensje)
odchylenie = pensje.std()

print(f'Średnia pensja : {średnia} zł')
print(f'Mediana pensja:{mediana} zł')
print(f'Odchylenie standartowe: {odchylenie} zł')
print(f'Pracownicy z pensją wyżej średniej {np.array(pracownicy) [pensje > średnia]}')
print(f' Pracownicy z pensją niżej mediany : {np.array(pracownicy)[pensje<mediana]}')
różnica = pensje.max() - pensje.min()
iqr= np.percentile(pensje, 75) - np.percentile(pensje, 25)
print(f'Różnica :{różnica} zł')
print(f'IQR: {iqr} zł')
```

Zadanie 3b

```python
premie= np.where(oceny>= 4.5, 2000, np.where(oceny >=4.0, 1000,0))
nowe_wynagrodzenia= pensje+ premie
najwyzsza_premia_idx=premie.argmax()
print(f'Naliczone premie: {premie}')
print('Całe wynagrodzenie: {nowe_wynagrodzenie}')
print(f'Nawiększą premię otzymała {pracownicy[najwyzsza_premia_idx]}')
```

Zadanie 3c

```python
kolejność = np.argsort(pensje)[::-1]
print('Kolejność pracowników zgodnie z  premią:')
for i, idx in enumerate(kolejność, 1):
    print(f'{i}.{pracownicy[idx]} Pensja: {pensje[idx]} zł, Staż :{staz[idx]} lat')
```

Zadanie 3d

```python
kor_staz_pensja = np.corrcoef(staz, pensje)[0,1]
kor_staz_ocena= np.corrcoef(staz, oceny)[0,1]
print(f'Korelacje staz/ pensja {kor_staz_pensja}')
print(f'Korelacje staz/ocena {kor_staz_ocena}')
if abs(kor_staz_pensja)> abs(kor_staz_ocena):
    print('Korelacka między stażem i pensję jest mocniejsza')
else:
    print('Korelacja między stażem i oceną jest mocniejsza')    
```

Zadanie 3e

```python
ceny_kol = ceny_produktow.reshape(4,1)
przychod = sprzedaz * ceny_kol
najlepszy_produkt_idx = przychod.sum(axis=1).argmax()
najlepszy_miesiac_idx = przychod.sum(axis=0).argmax()
pierwsze_3 = przychod[:,:3].sum()
ostatnie_3 = przychod[:,3:].sum()
print(f'Najwiekszy przychod od produktu: {produkty_nazwy[najlepszy_produkt_idx]}')
print(f'Najlepszy miesiąc łącznie {najlepszy_miesiac_idx + 1}')
print(f'Zarobki za piersze trzy miesiące:{pierwsze_3} zł')
print(f'Zarobki za ostatnie 3 niesiące:{ostatnie_3} zł')
if ostatnie_3>pierwsze_3:
    print('Trend rosnący')
else:
    print('Trend malejący')
```

```python
np.random.seed(42)
wiek=np.clip(np.random.normal(35,10,100).astype(int), 18, None)
wydatki = np.clip(np.random.normal(500,200,100), 0, None)
wizyty =np.random.poisson(5,100)
print('Statystyki opisowe')
print(f'Wiek -> średnia:{wiek.mean():.1f}, mediana: {np.median(wiek):.1f}, std:{wiek.std():.1f}, min: {wiek.min()}, max:{wiek.max()}')
print(f'Wydatki -> średnia:{wydatki.mean():.1f}, mediana: {np.median(wydatki):.1f}, std:{wydatki.std():.1f}, min: {wydatki.min()}, max:{wydatki.max()}')
print(f'Wizyty -> średnia:{wizyty.mean():.1f}, mediana: {np.median(wizyty):.1f}, std:{wizyty.std():.1f}, min: {wizyty.min()}, max:{wizyty.max()}')
print('Korelacje')
kor_wiek_wydatki= np.corrcoef(wiek,wydatki)[0,1]
kor_wizyty_wydatki = np.corrcoef(wizyty, wydatki)[0,1]
print(f'Korelacja wiek i wydatki:{kor_wiek_wydatki:.3f}')
print(f'Korelacjawizyty i wydatki:{kor_wizyty_wydatki:.3f}')
sr = wydatki.mean()
sd= wydatki.std()
warunki= [wydatki > (sr+sd), wydatki<(sr-sd)]
wybory= ['Wysoki', 'Niski']
segmenty = np.select(warunki, wybory, default = 'Średni')
print('Wyniki')
unikalne_segmenty, liczby_klientów= np.unique(segmenty, return_counts=True)
for segment, liczba in zip(unikalne_segmenty, liczby_klientów):
    procent=(liczba/len(segmenty)) *100
    print(f'Segment {segment}:{liczba} klientów ({procent:.0f})%')
```
