Ćwiczenie 1

Zadanie 1a

```python
df.info()
braki = df.isna().sum()
print('\nSuma braków :', braki)
procent_brakow = (df.isna().sum()/ len(df) * 100).round(1)
print('\n Procent braków', procent_brakow)
wiersze_z_nan = df[df.isna().any(axis = 1)]
print('\n Wiersze z brakami', wiersze_z_nan)
```

Zadanie 1b

```python
print("Unikalne w wynagrodzeniu:", df['wynagrodzenie'].unique())
df['wynagrodzenie'] = df['wynagrodzenie'].replace('brak', np.nan)
df['wynagrodzenie'] = pd.to_numeric(df['wynagrodzenie'], errors = 'coerce')
print('Liczba nan w wynagrodzeniu:', df['wynagrodzenie'].isna().sum())
```

Zadanie 1c

```python
mediana = df['wynagrodzenie'].median()
print(f'Mediana: {mediana}')
df['wynagrodzenie'] = df['wynagrodzenie'].fillna(mediana)
srednia = round(df['ocena_roczna'].mean(), 2)
print(f'Średnia ocena: {srednia}')
df['ocena_roczna'] = df ['ocena_roczna'].fillna(srednia)
print('\n Pozostałe braki:', df.isna().sum())
```

Zadanie 1d

```python
df_bez_brakow = df.dropna(subset=['data_zatrudnienia'])
print(f'Liczba wierszy przed: {len(df)}')
print(f'Liczba wierszy po usunięciu braków: {len(df_bez_brakow)}')
```
