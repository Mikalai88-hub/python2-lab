Ćwiczenie 3

```python
# Produkty w sklepie internetowym
nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka',
         'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka']

ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])
```

zadanie 1

```python
srednia_cena = ceny.mean()
indeks_max =ceny.argmax()
indeks_min=  ceny.argmin()
print(f'Średnia cena:{srednia_cena:.2f} zł')
print(f'Najdroższy product:{indeks_max} ({ceny[indeks_max]}) zł')
print(f'Najtańszy product:{indeks_min} ({ceny[indeks_min]}) zł')
```

zadanie 2

```python
wartość_produktów = ceny* ilosci_magazyn
wartość_magazynu= wartość_produktów.sum()
print (f'Wartość każdego produkta: {wartość_produktów} zł')
print(f'Wartość magazynu:{wartość_magazynu} zł')
```

Zadanie 3

```python
rabat= ceny * 0.85
VAT= ceny * 1.23
Drożej_1000 = (ceny>1000).sum()
print(f'Cena z promocją:{rabat} zł')
print(f'Ceny z VAT: {VAT} zł')
print(f'Ilość produktów powyżej 1000 zł: {Drożej_1000}')
```

Zadanie 4

```python
maska_oceny=oceny_klientow >=4.5
print(f'Produkty z oceną >=4.5: Nazwy:{np.array(nazwy)[maska_oceny]}\n Ceny: {ceny[maska_oceny]}')
maska_ceny= ceny < 200
print(f'Produkty poniżej 200 zł: Nazwy:{np.array(nazwy)[maska_ceny]}\n Ilości: {ilosci_magazyn[maska_ceny]}')
```

Zadanie 5

```python
maska_zapas = ilosci_magazyn < 30
wartosc_malo = wartość_produktów[maska_zapas].sum()
procent_magazynu= (wartosc_malo/wartość_magazynu) * 100
print(f'Produkty < 30 na stanie :{np.array(nazwy)[maska_zapas]}')
print(f'Łączna wartosc tyvh produktow:{wartosc_malo}')
print(f'Stanowią one {procent_magazynu:.2f}% magazynu')
```

```python
maska_okazja = (oceny_klientow >=4.5)& (ceny < 1000)
print(f'Specjalne okazje : {np.array(nazwy)[maska_okazja]}\n Ceny:{ceny[maska_okazja]}')
średnia1=ceny[oceny_klientow>=4.0].mean()
średnia2=ceny[oceny_klientow<4.0].mean()
print(f'Średnia cena produktów w oceną powyżej 4.0: {średnia1:.2f} zł')
print(f'Średnia cena produktów w oceną poniżej 4.0: {średnia2} zł')
```

Ćwiczenie 4


```python
import time

# Przygotowanie danych
rozmiar = 1_000_000
lista = list(range(rozmiar))
tablica = np.array(lista)

# Test 1: Pomnóż przez 2
start = time.perf_counter()
wynik_lista = [x * 2 for x in lista]
czas_lista = time.perf_counter() - start

start = time.perf_counter()
wynik_numpy = tablica * 2
czas_numpy = time.perf_counter() - start

print(f"=== Mnożenie ×2 ({rozmiar:,} elementów) ===")
print(f"Lista:  {czas_lista*1000:.1f} ms")
print(f"NumPy:  {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
# Test 2: Suma
start = time.perf_counter()
suma_lista = sum(lista)
czas_lista = time.perf_counter() - start

start = time.perf_counter()
suma_numpy = tablica.sum()
czas_numpy = time.perf_counter() - start

print(f"\n=== Suma ({rozmiar:,} elementów) ===")
print(f"Lista:  {czas_lista*1000:.1f} ms")
print(f"NumPy:  {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
# Test 3: Filtrowanie (elementy > 500000)
start = time.perf_counter()
filtr_lista = [x for x in lista if x > 500000]
czas_lista = time.perf_counter() - start

start = time.perf_counter()
filtr_numpy = tablica[tablica > 500000]
czas_numpy = time.perf_counter() - start

print(f"\n=== Filtrowanie > 500k ({rozmiar:,} elementów) ===")
print(f"Lista:  {czas_lista*1000:.1f} ms")
print(f"NumPy:  {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

## Wnioski z benchmarku
- NumPy jest 21x szybszy w mnożeniu

- NumPy jest 27x szybszy w filtrowaniu
