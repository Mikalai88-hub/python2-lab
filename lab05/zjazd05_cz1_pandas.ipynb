```python
dane_placowe = {'Anna': 5500, 'Jan':7200,'Ewa':4800,'Marek':9300, 'Kasia':6300}
pensje= pd.Series(dane_placowe)
```

Zadanie 1b

```python
najbogatszy = pensje.idxmax()
print(f'Najwięcej zarabia: {najbogatszy}')
srednia= pensje.mean()
print(f'Srednia pensja:{srednia:.2f} zł')
powyzej_sredniej= pensje[pensje>srednia]
print('Osoby zarabiające powyżej sredniej')
print(powyzej_sredniej.index.tolist())
ponizej_6000 = (pensje < 6000).sum()
print(f'Liczba osób zarabiających poniżej 6000 :{ponizej_6000} ')
```

Zadanie 1c

```python
podwyzka_10 = pensje *1.10
premia_500 = pensje + 500
nowe_pensje = (pensje *1.10)+500
print('\nNowe pensje (podwyżka 10% + 500 zł premii):')
print(nowe_pensje)
ranking = nowe_pensje.sort_values(ascending = False)
print('\nRanking pensji (od najwyższej):')
print(ranking)
```

```python
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')

# Dodaj kolumnę z procentem napiwku
tips['tip_pct'] = (tips['tip'] / tips['total_bill'] * 100).round(1)
```

```python
sredni_rachunek= tips['total_bill'].mean()
print(f'Sredni rachunek: {sredni_rachunek:.2f} $')

sredni_tip_pct = tips['tip_pct'].mean()
print(f'Sredni napiwek:{sredni_tip_pct:.1f} %')

najczestszy_dzien=tips['day'].value_counts().idxmax()
print(f'Dzień z najwięksą liczbą rachunków: {najczestszy_dzien}')

palacze_count = tips ['smoker'].value_counts()
print(f'rachunki od palaczy: {palacze_count}')

print('Statystyki dla kolumny tips')
print(tips['tip'].describe())
print(f'Mediana napiwku: {tips['tip'].median()}$')

procent_czasu = tips['time'].value_counts(normalize=True) * 100
print(f'Udział procentowy pór dnia: {procent_czasu}')
print(f'Procent rachunków "Dinner":{procent_czasu['Dinner']:.1f} %')

unikalne_grupy = tips['size'].nunique()
print(f'\nLiczba unikalnych rozmirów grup:{unikalne_grupy}')
```
